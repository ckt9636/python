## 3. [스크래핑] https://www.timeanddate.com/holidays/south-korea/ 사이트에서 화면에 보이는 테이블의 holiday정보를 스크래핑 하여 아래 이미지와 동일한 컬럼의 정보를 추출한 후 KOPO_PY_{이름}_FINAL3 테이블로 생성하세요 [심심하면 Date도 같이..저장하세요]

```python
import requests, bs4, pandas as pd, numpy as np

# 웹페이지 html 소스 가져오기
resp = requests.get("https://www.timeanddate.com/holidays/south-korea/")
#resp.encoding='utf-8'
html = resp.text

bs = bs4.BeautifulSoup(html, 'html.parser')
holiday_list = bs.find("table", {"class":"zebra fw tb-cl tb-hover"}).select("tbody > tr")
holiday_list
```

```python
holiday_dateList = []
holiday_weekList = []
holiday_nameList = []


holiday_typeList1=[]
for content in holiday_list:
    alnfos=content.findAll("td")
    for i in alnfos:
        result=i.text
        holiday_typeList1.append(result)

holiday_typeList2=[]
for i in holiday_typeList1:
    if holiday_typeList1.index(i)%3 ==2:
        result=i
        holiday_typeList2.append(result)
print(holiday_typeList2)

for holiday_one in holiday_list:    
    holiday_date = holiday_one.find("th").text
    holiday_week = holiday_one.find("td",{"class":"nw"}).text   
    holiday_name = holiday_one.find("a").text   #.find{'a'}.text
    
    
    holiday_dateList.append(holiday_date)
    holiday_weekList.append(holiday_week)
    holiday_nameList.append(holiday_name)
    
    
        
list1 = pd.DataFrame(holiday_dateList)
holiday_weekList
list2 = pd.DataFrame(holiday_weekList)
holiday_nameList
list3 = pd.DataFrame(holiday_nameList)
holiday_typeList
list4 = pd.DataFrame(holiday_typeList2)
final = pd.concat([list1, list2, list3, list4], axis = 1)
final.columns=['DATE', 'WEEK', 'HOLIDAY_NAME', 'HOLIDAY_TYPE']
final
```

```python
# 데이터 저장
resultname='KOPO_PY_최경재_FINAL3'
final.to_sql(resultname, engine, if_exists='replace', index=False)
```
